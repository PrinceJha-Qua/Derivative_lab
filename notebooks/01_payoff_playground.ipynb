```python
from payoff import long_call, long_put, short_call, short_put , long_call_profit, long_put_profit, short_call_profit, short_put_profit, bull_call_spread
```

```python
S = np.arange(50, 251)
K = 150

plt.plot(S, long_call(K, S), label="Long Call")
plt.plot(S, short_call(K, S), label="Short Call")

plt.plot(S, long_put(K, S), label="Long Put")
plt.plot(S, short_put(K, S), label="Short Put")

plt.axhline(0, linestyle="--")
plt.axvline(K, linestyle="--")

plt.legend()

plt.xlabel("Stock Price")
plt.ylabel("Payoff")

plt.title("Vanilla Option Payoffs")

plt.grid(True)

plt.show()
```

### Observation
- Long positions exhibit convex upside exposure while short positions inherit convex downside risk.
- The strike price acts as the transition boundary between active and inactive payoff regions.
- Short options are not simply “inverse investments” psychologically, they expose the trader to potentially asymmetric risk.

```python
S = np.arange(50, 251)
K = 150
premium = 8


plt.plot(S, long_call_profit(K, S, premium), label="Long Call")
plt.plot(S, short_call_profit(K, S, premium), label="Short Call")

plt.plot(S, long_put_profit(K, S, premium), label="Long Put")
plt.plot(S, short_put_profit(K, S, premium), label="Short Put")

plt.axhline(0, linestyle="--")
plt.axvline(K, linestyle="--")

plt.legend()

plt.xlabel("Stock Price")
plt.ylabel("Profit")

plt.title("Vanilla Option Profits")

plt.grid(True)

plt.show()
```

## Observation

A positive payoff does not necessarily imply profit.

The premium shifts the entire payoff structure downward, creating a breakeven point beyond the strike price.
